ndim() returns the number of dimension of the data struture. 
Axes() returns a list of the axes of the row labels

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('MathScoreTerm1.csv')
# print(df)
print('The Number of Dimension: ', df.ndim)
print(df.axes)

```

'std()' returns the standard deviation of the values for the requested axis

```python
print(df['Age'].std())
```

iteritems() iterates over each column as key and value as pair

```python
for key, value in df.iteritems():
    print(key,value)
```

itertuples() returns as a named tupple for each row

```python
for row in df.itertuples():
    print(row)
```

```python
df_phy = pd.read_csv('PhysicsScoreTerm1.csv')
df_concat=pd.concat([df,df_phy])
print(df_concat)
# print(df.append(df_phy))
```
