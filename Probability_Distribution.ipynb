```python
m = np.random.randint(2,10,40)
m
```

```python
df = pd.DataFrame(m)
df= pd.DataFrame(df[0].value_counts())
df
```

```python
length = len(m)
length
```

```python
df.columns = ['Counts'] 
df['Prob'] = df['Counts']/length
df
```

```python
sns.barplot(df['Counts'],df['Prob'])
```

### Probability Distribution Function 

```python
mu = 20 
sigma = 2
h = sorted(np.random.normal(mu,sigma,100))
```

```python
fit = stats.norm.pdf(h,np.mean(h),np.std(h))
plt.plot(h,fit,'-o')
plt.hist(h,normed=True)
```

### Cumalative Distribution Function

```python
sfit =  stats.norm.cdf(h,np.mean(h),np.std(h))
```

```python
plt.plot(h,fit, label = 'pdf')
plt.plot(h,sfit, label = 'cdf')
```
