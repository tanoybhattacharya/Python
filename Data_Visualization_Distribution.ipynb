SINGLE LINE PLOT 

```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
import pandas as pd
from IPython.display import Image


plt.plot([1,2,3,4,11])
plt.show()
```

```python
x=[1,2,3,4,10]
plt.plot(x,[y**2 for y in x]) # Here x axix = [1,2,3,4,10] and Y axis = x*x
plt.show()
```

```python
x = range(5)
plt.plot(x,[y**3 for y in x]) # x Axix = [0,1,2,3,4] and Y axix =x*x*x
plt.show()
```

```python
x= np.arange(0,5,0.1)
plt.plot(x,[y**2 for y in x])
plt.show()
```

MULTILINE PLOT

```python
x = np.arange(1,6,0.5)
# Three line are plotted
plt.plot(x,[y for y in x ])
plt.plot(x,[y**2 for y in x ])
plt.plot(x,[y**3 for y in x ])
plt.show()
```

```python
x = np.arange(0,5,.5)
plt.grid(True) # show teh grid line
plt.plot(x,[y for y in x],x,[y**2 for y in x],x,[y**3 for y in x])
plt.show()
```

ADDING LABEL IN THE PLOT

```python
x = np.arange(0,5,.5)
plt.grid(True) # show teh grid line
plt.plot(x,[y for y in x],x,[y**2 for y in x],x,[y**3 for y in x])
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')
plt.title('Learing Matplotlib')
plt.show()
```

ADDING LEGENDS

```python
x = np.arange(1,6,0.5)
# Three line are plotted
plt.plot(x,[y for y in x ], label='Linear')
plt.plot(x,[y**2 for y in x ], label='Square')
plt.plot(x,[y**3 for y in x ],  label='Qube')
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')
plt.title('Learing Matplotlib')
plt.grid(True)
plt.legend()
plt.show()
```

SAVING GRAPH AS IMAGE

```python
x = np.arange(1,6,0.5)
# Three line are plotted
plt.plot(x,[y for y in x ], label='Linear')
plt.plot(x,[y**2 for y in x ], label='Square')
plt.plot(x,[y**3 for y in x ],  label='Qube')
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')
plt.title('Learing Matplotlib')
plt.grid(True)
plt.legend()
plt.savefig('plot.png')
```

CREATE BAR CHART USING DICTIONARIES

```python
dict ={'A':45,'B':23,'C':67,'D':98}
for i,key in enumerate(dict): #enumerate(dict) create [(0,'A'),(1,'B'),(2,'C'),(3,'D')] 
    plt.bar(i,dict[key])
plt.xticks(np.arange(len(dict)),dict.keys()) #Adds the keys as lebel on the axis
# plt.xticks([0,1,2,3],dict.keys()) # Adds the keys as lebel on the axis
plt.show()
```

STLYING GRAPH

```python
y = np.arange(1,3)

plt.plot(y,'y') # Second Parameter provide the color code which support by matplotlib
plt.plot(y+5,'m') #[b:blue,c:cyan,g:green,k:black,m:Magenta,r:Red,w:White,y:Yellow]
plt.plot(y+10,'c')
plt.show()
```

```python
y = np.arange(1,3)

plt.plot(y,'--') 
plt.plot(y+5,'r') 
plt.plot(y+10,':') # Second Parameter style the line
plt.show()
```
