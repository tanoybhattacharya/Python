<b>Web Scraping </b> is a technique employed to extract large amount of data from websites whereby the data is extracted and save to alocal file in your computer or to a database

<b>Beautiful Soup</b> is a python library for pulling data out of HTML and XML files.

<b>Request</b> is a python library that is widely used for sending and receiving information over HTTP

```python
import requests
from bs4 import BeautifulSoup
```

```python
r = requests.get('https://www.edureka.co/')
c= r.content
soup = BeautifulSoup(c,'html.parser')
# soup.prettify() # BeautifulSoup's prettify() is used to print the HTML content in the tree structure
```

Extracting URL from the webpage

```python
for link in soup.find_all('a'):
    print(link.get('href'))
```

Saving Extraticting Data in a list 

```python
links = []
for link in soup.find_all('a'):
    links+=[link.get('href')]
links
```
