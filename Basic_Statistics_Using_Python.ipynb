## Inferential Statistics ##

__Infer the properties of Population base on sample__

```python
population = np.random.randint(10,20,1000)

estimate = []

for x in range(200): # Generate 200 sample
    sample = np.random.choice(population,100)
    estimate.append(sample.mean())
```

```python
np.mean(estimate)
```

```python
pd.DataFrame(estimate).plot(kind='density')
plt.grid(True)
```

```python
population.mean()
```

### Point of Estimate

__Confidence Intervel__: (Sample_Mean - Mergin_Of_Error,Sample_Mean + Mergin_Of_Error)

__Mergin _of _Error:__ criticalvale*(standard deviation/sqrt(sample_size))

```python
Image(filename='CI.jpg')
```

```python
z_critical= stats.norm.ppf(0.95) # 95% CI level for Z value
```

```python
t_critical = stats.t.ppf(q=0.95,df=24) #determine T Value for 95% CI level and 25 sample size (degree of freedom = n-1)
t_critical
```

```python
mergin_of_error = z_critical*(np.std(estimate)/np.sqrt(200))
mergin_of_error
```

```python
#Lower limit

np.mean(estimate) - mergin_of_error
```

```python
#Upper limit

np.mean(estimate) + mergin_of_error
```

__Inference__: We came to colusion based on sample that 95% chance population mean will be in between 14.45 and 14.52. Although we know in this example population mean 14.59.

#### Continuous Variable ####

```python
mu = 20 # Mean
sigma = 2 # Stand Deviation
h = sorted(np.random.normal(mu,sigma,100)) # create random sample from a normal distribution

# vaerification
abs(20-np.mean(h)) < 2
```

```python
plt.hist(h,30,normed=True) # Bin size 30
```

```python
fit = stats.norm.pdf(h,np.mean(h),np.std(h))
plt.plot(h,fit,'-o')
plt.hist(h,normed=True)
```
